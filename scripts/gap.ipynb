```python
import os
from PIL import Image
from pathlib import Path
import shutil
import numpy as np
```

```python
in_folder = Path("datasets/new_broken_masks")
out_folder = Path("datasets/skill_lack")
```

```python
from tqdm import tqdm
list = sorted(Path(in_folder).glob("**/*.jpg"))
shutil.rmtree(out_folder, ignore_errors=True)

for _, file in tqdm(enumerate(list)):
    # png = Path(str(file).replace(".jpg", "-masklabel.png"))
    
    original = np.array(Image.open(file).convert("L"), dtype=np.uint8)
    # mask = np.array(Image.open(png).convert("L"))
    # mask_bool = mask > 0
    noise = np.random.randint(0, 256, size=original.shape, dtype=np.uint8)
    # new = original.copy()
    # new[mask_bool] = noise[mask_bool]


    os.makedirs(out_folder.joinpath(*file.parts[2:-1]), exist_ok=True)
    loc = out_folder.joinpath(*file.parts[2:-1]) / file.with_suffix(".jpg").name
    Image.fromarray(noise).save(loc)
```
