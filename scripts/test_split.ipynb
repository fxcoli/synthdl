```python
import os
import shutil
from pathlib import Path
import pandas as pd
import math
```

```python
data = pd.read_csv("all.csv")
broken = data.loc[data['normal'] == 0][data.columns[1:-1]].sample(frac=1)   # shuffles

count_dict = {col: math.floor(broken[col].loc[broken[col] == 1].count() * 0.3) for col in broken.columns[1:]}
print(count_dict)
id_list = []

for _, row in broken.iterrows():
    id = row['ID'][:-4]
    add = True
    for col in row.keys()[1:]:
        if row[col] == 1 and count_dict[col] == 0:  # If adding would overflow the pathologies requirement
            add = False

    if add:
        id_list.append(id)
        for col in row.keys()[1:]:
            if row[col] == 1:
                count_dict[col] -= 1

id_list
```

```python
in_folder = Path("datasets/new_broken_masks")
out_folder = Path("datasets/test_broken")
```

```python
from tqdm import tqdm
shutil.rmtree(out_folder, ignore_errors=True)
os.makedirs(out_folder)

for _, file in tqdm(enumerate(id_list)):
    dupes = sorted(Path(in_folder).glob(f"**/{file}*"))

    for dupe in dupes:
        loc = out_folder.joinpath(*dupe.parts[2:-1])
        os.makedirs(loc, exist_ok=True)
        shutil.move(dupe, loc / dupe.name)
```
