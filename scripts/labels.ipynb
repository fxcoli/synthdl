```python
from pathlib import Path
import pandas as pd
```

```python
train = Path("../datasets/train")
train_files = list(train.rglob("*.jpg"))
csv = pd.read_csv("../datasets/all.csv")
csv
```

```python
def make_label(file: Path):
    loc = file.parent
    name = file.name
    newloc = loc / (file.stem + ".txt")
    try:
        row = csv[csv["ID"] == name].iloc[0]
    except Exception as e:
        print(e)
        print(file)
        return

    label = "An anteroposterior " if "AP" in name else "A lateral "
    label += "radiograph of a"
    if row["normal"] == 1:
        label += " healthy elbow"
    else:
        label += "n elbow displaying " + ', '.join(row[row == 1].index.tolist())

    with open(newloc, 'w') as lf:
        lf.write(label)

l = [make_label(file) for file in train_files]
```
